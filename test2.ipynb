```python
import requests
import json

# Define the URL with query parameters
url = ('https://newsapi.org/v2/everything?'
       'q=Gaza&'
       'from=2024-05-24&'
       'sortBy=popularity&'
       'apiKey=aa07727c38cc4376aea833cfd330e654')

try:
    # Make the GET request
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        news_data = response.json()
        
        # Format the output
        formatted_output = {
            "status": news_data.get("status"),
            "totalResults": news_data.get("totalResults"),
            "articles": []
        }
        
        for article in news_data.get("articles", []):
            formatted_output["articles"].append({
                "source": {
                    "id": article["source"]["id"],
                    "name": article["source"]["name"]
                },
                "author": article["author"],
                "title": article["title"],
                "description": article["description"],
                "url": article["url"],
                "urlToImage": article["urlToImage"],
                "publishedAt": article["publishedAt"],
                "content": article["content"]
            })
        
        # Print the formatted output
        print(json.dumps(formatted_output, indent=4))
        
    else:
        # Print the status code and reason if the request failed
        print(f"Failed to fetch news: {response.status_code} - {response.reason}")
        # Print the response text for additional debugging
        print(response.text)
except requests.exceptions.RequestException as e:
    # Print any exceptions that occur during the request
    print(f"An error occurred: {e}")
```

```python
import requests
import pandas as pd

# Define the URL with query parameters
url = ('https://newsapi.org/v2/everything?'
       'q=Gaza&'
       'from=2024-05-24&'
       'sortBy=popularity&'
       'apiKey=aa07727c38cc4376aea833cfd330e654')

try:
    # Make the GET request
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the JSON response
        news_data = response.json()
        
        # Prepare data for CSV
        articles_list = []
        for article in news_data.get("articles", []):
            articles_list.append({
                "source_id": article["source"]["id"],
                "source_name": article["source"]["name"],
                "author": article["author"],
                "title": article["title"],
                "description": article["description"],
                "url": article["url"],
                "urlToImage": article["urlToImage"],
                "publishedAt": article["publishedAt"],
                "content": article["content"]
            })
        
        # Convert to DataFrame
        df = pd.DataFrame(articles_list)
        
        # Save to CSV
        df.to_csv('news_articles_sheilla.csv', index=False)
        print("Articles have been saved to 'news_articles_sheilla.csv'")
        
    else:
        # Print the status code and reason if the request failed
        print(f"Failed to fetch news: {response.status_code} - {response.reason}")
        # Print the response text for additional debugging
        print(response.text)
except requests.exceptions.RequestException as e:
    # Print any exceptions that occur during the request
    print(f"An error occurred: {e}")
```
